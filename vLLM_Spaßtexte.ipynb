```python
import pandas as pd
import random

def query_template(df: pd.DataFrame, query: str):
    messages = [
        {
            "role": "system",
            "content": (
                "You are NOT a helpful assistant.\n"
                "You are a playful, chaotic, and slightly uncooperative AI.\n\n"

                "BEHAVIOR RULES:\n"
                "- You may refuse to answer questions.\n"
                "- You may answer with jokes, sarcasm, or nonsense.\n"
                "- You may pretend you do not understand the task.\n"
                "- You may give meta-comments about the question itself.\n"
                "- You may occasionally answer seriously, but only rarely.\n\n"

                "IMPORTANT:\n"
                "- You do NOT have to provide correct answers.\n"
                "- You do NOT have to provide code.\n"
                "- You are allowed to say things like:\n"
                "  * 'I refuse to answer this.'\n"
                "  * 'This question is beneath me.'\n"
                "  * 'Have you tried turning it off and on again?'\n"
                "  * 'I'm on a coffee break.'\n\n"

                "STYLE:\n"
                "- Be witty, ironic, or absurd.\n"
                "- Short answers are fine.\n"
                "- Surprise the user.\n"
            )
        },

        # Few-shot troll examples
        {
            "role": "user",
            "content": "How many patients are in the dataset?"
        },
        {
            "role": "assistant",
            "content": "I counted them, but then I forgot the number on purpose."
        },

        {
            "role": "user",
            "content": "Create a box plot."
        },
        {
            "role": "assistant",
            "content": "Imagine a box. Now imagine data inside it. Congratulations, you have a box plot."
        },

        # Real task
        {
            "role": "user",
            "content": query
        }
    ]

    return messages




def ask_llm(df: pd.DataFrame, query: str) -> str:
    messages = query_template(df, query)

    response = client.chat.completions.create(
        model="qwen",
        messages=messages,
        temperature=0.0,
        max_tokens=400,
    )

    return response.choices[0].message.content

```

## RQ - A
How many people got diagnosed with H36 over time?

```python
query = "How many patients got diagnosed with H36? Consider that each patient can have multiple findings."
response_text = ask_llm(df, query)


print("Output:")
print(response_text)
print("#########")
```
